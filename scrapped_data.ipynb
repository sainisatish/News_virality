```python
import requests
import pandas as pd
from bs4 import BeautifulSoup ,SoupStrainer
```

### News Articles is scraped from HindustanTimes and it consist two type of news genuine and most viral news

```python
genuine_news_url="https://www.hindustantimes.com/india-news/"
viral_news_url='https://www.hindustantimes.com/it-s-viral/'
```

## Extracting Genuine News

```python
# filter for extracting target data
def largetext(tags):
    return True if tags.name=='a' and len(tags.get_text())>=50 else False
```

```python
def scrape_data(url,tag):
    req=requests.get(url)
    if req.status_code==200:
        content=req.content
        tags_to_filter=SoupStrainer(tag)
        soup_obj=BeautifulSoup(content,'html.parser',parse_only=tags_to_filter)
        
        return soup_obj

    else:
        false      
```

```python
obj=scrape_data(genuine_news_url,'a')
```

```python
news=obj.find_all(largetext)
```

```python
len(news)
```

```python
def largetext(tags):
    return True if tags.name=='p' and len(tags.get_text())>=100 else False
```

```python
list_of_summary=[]
summary=' '
for i in range(len(news)):
    obj=scrape_data(news[i]['href'],'p')
    summary_obj=obj.find_all(largetext)
    for j in range(1,len(summary_obj)):
        summary+=summary_obj[j].get_text()
    list_of_summary.append(summary)
    summary=' '
    print("paragraph extracted: ", i)
    
```
